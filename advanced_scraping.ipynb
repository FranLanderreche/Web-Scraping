```python
import json
data= json.loads(json_data)
```

```python
def get_page(collection, page):
    print "requesting page %d" % page
    req = mechanize.Request("http://www.caprally.com/load_investor_list", "page=%d&search_box=&sector=all&stage=all&geo=United+States&latitude=&longitude=&zipcode=" % page)
    
    for header in parse_head:
        req.add_header(header[0], header[1])
    resp = br.open(req).read()
    j_data= json.loads(resp)
    collection += j_data['investors']
    return collection
```
